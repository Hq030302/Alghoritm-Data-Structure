```python
import pandas as pd
import numpy as np
import string
import re
import nltk
import Sastrawi
```

```python
#tweet_okt = pd.read_csv("filtered_tweets_okt.csv")
tweet_aug_sept = pd.read_csv("filtered_tweets_aug-sept.csv")
#print(tweet_okt)
print(tweet_aug_sept)
```

```python
tweet = tweet_aug_sept['full_text']
```

CASE FOLDING & Cleaning Text

```python
def clean_text(text):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).split())
```

```python
tweet_aug_sept['clean'] = tweet_aug_sept.apply(lambda x: clean_text(x['full_text']), axis=1)
tweet = tweet_aug_sept['clean']
```

```python
tweet_aug_sept.head(10)
tweet_aug_sept.to_csv('clean_tweets_aug_sept.csv', index=True)
```

TOKENIZING

```python
from nltk.tokenize import RegexpTokenizer
regexp = RegexpTokenizer(r'\w+|$[0-9]+|\S+')
tweet_aug_sept['Token'] = tweet_aug_sept['clean'].apply(regexp.tokenize)
```

```python
tweet_aug_sept
tweet_aug_sept.to_csv('tokenized_tweets_aug-sept.csv', index=True)
```

Stopword Removal

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()
print(stopwords)
```

```python
def stopwords_text(tokens):
    cleaned_tokens = []
    for token in tokens:
        if token not in stopwords:
            cleaned_tokens.append(token)
    return cleaned_tokens
```

```python
tweet_aug_sept['stop'] = tweet_aug_sept['Token'].apply(stopwords_text)
tweet_aug_sept.head(10)
```

STEMMING

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
stem_factory = StemmerFactory()
stemmer = stem_factory.create_stemmer()
```

```python
def stemming_text(tokens):
  hasil = [stemmer.stem(token) for token in tokens]
  return hasil
```

```python
tweet_aug_sept['stemmed'] = tweet_aug_sept['stop'].apply(stemming_text)
tweet_aug_sept.head()
```
